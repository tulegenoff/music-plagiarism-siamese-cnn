```python
def write_file(requests, case_number, case_letter, path):
    filename = a['href'][a['href'].rfind("/")+1:]
    doc = requests.get(a['href'])
    print(filename)
    print(f'case{case_number}_{case_letter}.mp3')
    print('---------------------')
    with open(os.path.join(f'{path}\\', f'case{case_number}_{case_letter}.mp3'), 'wb') as f:
        f.write(doc.content)
        #f.write(AudioSegment.from_mp3(doc.content).export(os.path.join('non_plagiarized/', f'case{1}_a.wav'), format='wav'))
    my_file = Path(f"C:\\Users\\Tulegenoff\\PetProjects\\Diploma\\{path}\\" + f'case{case_number}_{case_letter}.mp3')
    AudioSegment.from_mp3(my_file).export(Path(f"C:\\Users\\Tulegenoff\\PetProjects\\Diploma\\{path}\\" + f"{f'case{case_number}_{case_letter}.mp3'[:-4]}.wav"), format="wav")
    os.remove(os.path.join(f'{path}/', f'case{case_number}_{case_letter}.mp3'))

def write_file_wav(requests, case_number, case_letter, path):
    filename = a['href'][a['href'].rfind("/")+1:]
    doc = requests.get(a['href'])
    print(filename)
    print(f'case{case_number}_{case_letter}.mp3')
    print('---------------------')
    with open(os.path.join(f'{path}\\', f'case{case_number}_{case_letter}.wav'), 'wb') as f:
        f.write(doc.content)
      
        
```

```python
import re
import os
from time import sleep
import random
case_number = 1
headers = {'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, Like Gecko) Chrome/112.0.5615.138 Safari/537.36'}
for i in url_cases:
    try:
        r = requests.get(i, headers = headers)
        content = r.content
        soup = BeautifulSoup(content, 'html.parser')
        table = soup.find('table', id="tablepress-4")
        cnt = 1
        amount_of_files = len(soup.find_all('a', href=re.compile(r'http.*\.mp3')))  
        if(amount_of_files!=0 and amount_of_files==2):
            for a in soup.find_all('a', href=re.compile(r'http.*\.mp3')):
                if cnt%2!=0:
                    write_file(requests, case_number, 'a', 'non_plagiarized')
                else:
                    write_file(requests, case_number, 'b', 'plagiarized')
                cnt = cnt + 1
            case_number = case_number + 1
            sleep(random.randint(5,10))
        elif(amount_of_files!=0 and amount_of_files%2==0):
            for a in soup.find_all('a', href=re.compile(r'http.*\.mp3')):
                if cnt%2!=0:
                    write_file(requests, case_number, 'a', 'non_plagiarized')
                else:
                    write_file(requests, case_number, 'b', 'plagiarized')
                    case_number = case_number + 1
                cnt = cnt + 1
            sleep(random.randint(5,10))
                
    except:
        pass
```
